```python
"""
Simple Login System
Author : Sannidhya Bahulekar
Description : Checks whether entered Username and Password are correct
"""

# Predefined Credentials

username = "admin"
password = "1234"

# Input from User
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Check Credentials

if (input_username == username and input_password == password):
    print("Login Successful")
     
else:
    print("Invalid Credentials")


# End of Program
```

```python
"""
Pass/Fail Analyzer
Author : Sannidhya Bahulekar
Description : Analyzes student marks and count pass and fail students
"""
# List of Student Marks
marks = [45, 78, 90, 33, 66]
 
# Initialize Counter
pass_count = 0
fail_count = 0

# Loop through marks to evaluate results 

for mark in marks: 
    if mark >=50:
        pass_count += 1
    else:
        fail_count += 1


# Display final Results
print("Total Students:", len(marks))
print("Number of Students Passes:", pass_count)
print("Number of Students failed", fail_count)

# End Of Program
      
```

```python
"""
Simple Data Cleaner
Author : Sannidhya Bahulekar
Description: Cleans and standardizes user names.
"""

# List of names with inconsistent formatting
names = [" Alice ", "bob", "CHARLIE "]

# Clean names: it removes spaces and converts Uppercase to Lowercase
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Now Print The Clean List
print("Cleaned Names:", cleaned_names)

# End Of Program

```

```python
"""
Message Length Analyzer
Author : Sannidhya Bahulekar
Description: Analyzes message lengths and flags messages longer than 10 characters.
"""

# Given list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for message in messages:
    length = len(message)
    
    # Display message and its length
    print(f"Message: '{message}' | Length: {length}")
    
    # Check if message length is greater than 10
    if length > 10:
        print("This message is longer than 10 characters.\n")
    else:
        print("This message length is acceptable.\n")

# End of Program
```

```python
"""
ERROR Message Detector
Author : Sannidhya Bahulekar
Description : Counts the number of ERROR Entries in system logs
"""
# Given List of system log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counts number of ERROR messages
error_count = logs.count("ERROR")

# Print Result
print("Total number of ERROR messages detected:", error_count)

# End of Program
```
